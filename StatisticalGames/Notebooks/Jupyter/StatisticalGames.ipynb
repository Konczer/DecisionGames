# Statistical Games

This is an interactive notebook, reproducing plots and numerical calculations for Fisher, Bayesian, and Statistical games, based on the [arXiv:2402.15892](https://arxiv.org/abs/2402.15892) paper on [Statistical Games](https://arxiv.org/pdf/2402.15892):

> **Abstract:** This work contains the mathematical exploration of a few prototypical games in which central concepts from statistics and probability theory naturally emerge. The first two kinds of games are termed Fisher and Bayesian games, which are connected to Frequentist and Bayesian statistics, respectively. Later, a more general type of game is introduced, termed Statistical game, in which a further parameter, the players' relative risk aversion, can be set. In this work, we show that Fisher and Bayesian games can be viewed as limiting cases of Statistical games. Therefore, Statistical games can be viewed as a unified framework, incorporating both Frequentist and Bayesian statistics. Furthermore, a philosophical framework is (re-)presented -- often referred to as minimax regret criterion -- as a general approach to decision making.
<br>
The main motivation for this work was to embed Bayesian statistics into a broader decision-making framework, where, based on collected data, actions with consequences have to be made, which can be translated to utilities (or rewards/losses) of the decision-maker. The work starts with the simplest possible toy model, related to hypothesis testing and statistical inference. This choice has two main benefits: i.) it allows us to determine (conjecture) the behaviour of the equilibrium strategies in various limiting cases ii.) this way, we can introduce Statistical games without requiring additional stochastic parameters. The work contains game theoretical methods related to two-player, non-cooperative games to determine and prove equilibrium strategies of Fisher, Bayesian and Statistical games. It also relies on analytical tools for derivations concerning various limiting cases.

A simple function to check if the guessing player (Player 1 in the paper) can surely win the game.

````python
def is_surewinning(N, K_list, M):

    KA, KB = K_list

    k_A_minmax = [max(0, N - (M - KA)), min(N, KA)]
    k_B_minmax = [max(0, N - (M - KB)), min(N, KB)]

    k_minmax = [min(k_A_minmax[0], k_B_minmax[0]), max(k_A_minmax[1], k_B_minmax[1])]
    k_AB_minmax = [max(k_A_minmax[0], k_B_minmax[0]), min(k_A_minmax[1], k_B_minmax[1])]

    return k_AB_minmax[1] < k_AB_minmax[0]
````

Computing all equilibrium values for all $\mathfrak{Game}(N=4, K_A, K_B, M=10)$ games, and storing the results (grouped by equilibrium parameters $P, k, \nu, s, v$) in `results_dict`:

````python
# Set the parameters
M = 10
N = 4

# Set the ranges for KA and KB
KA_range = range(0, M + 1)
KB_range = range(0, M + 1)

# Initialize a matrix to hold dictionaries
results_matrix = np.empty((len(KA_range), len(KB_range)), dtype=dict)

# Compute equilibrium values for each combination of KA and KB
for i in KA_range:
    for j in KB_range:
        result = fishergame_solve(N, [i, j], M)
        results_matrix[i, j] = result

# Keys for results_dict
keys = list(results_matrix[0, 0].keys())

# Create a dictionary of matrices
results_dict = {key: np.zeros((len(KA_range), len(KB_range)), dtype=float) for key in keys}

# Populate the dictionary of matrices
for i, KA in enumerate(KA_range):
    for j, KB in enumerate(KB_range):
        for key in keys:
            results_dict[key][i, j] = results_matrix[i, j][key]
````

Setting a mask for "Sure winning" games:

````python
# mask for excluding sure winning games
mask = np.zeros((len(KA_range), len(KB_range)), dtype=bool)  # Create a boolean array

for i, KA in enumerate(KA_range):
    for j, KB in enumerate(KB_range):
        mask[i, j] = is_surewinning(N, [KA, KB], M)  # Assign the boolean result directly
````

## Bayesian game plots

Computing equilibrium values for all $\mathfrak{BGame}(N=4, K_A, K_B, M=10)$ games, and storing the results in `bayes_results_matrix`. (The values grouped by equilibrium paramaters $P^*, G^*$ are stored in `bayes_results_dict`):

````python
# Set the parameters
M = 10
N = 4

# Set the ranges for KA and KB
KA_range = range(0, M + 1)
KB_range = range(0, M + 1)

# Initialize a matrix to hold dictionaries
bayes_results_matrix = np.empty((len(KA_range), len(KB_range)), dtype=dict)

# Compute equilibrium values for each combination of KA and KB
for i in KA_range:
    for j in KB_range:
        bayes_result = bayesiangame_solve(N, [i, j], M)
        bayes_results_matrix[i, j] = bayes_result

# Keys for results_dict
bayes_keys = ['P', 'G']
#list(bayes_results_matrix[0, 0].keys())[]

# Create a dictionary of matrices
bayes_results_dict = {key: np.zeros((len(KA_range), len(KB_range)), dtype=float) for key in bayes_keys}

# Populate the dictionary of matrices
for i, KA in enumerate(KA_range):
    for j, KB in enumerate(KB_range):
        for key in bayes_keys:
            bayes_results_dict[key][i, j] = bayes_results_matrix[i, j][key]
````

Reproducing plots from [Figure 17](https://arxiv.org/pdf/2402.15892#figure.caption.72) in the paper, i.e. $P^*(K_A,K_B)$ and $\Delta G^*(K_A,K_B)$ values for $\mathfrak{BGame}(N = 4, K_A, K_B , M = 10)$ games:

$\Delta G^*(K_A, K_B)$:

The default Bayesian game in `bayesiangame_solve` is not a double or nothing, but rather a same or nothing, i.e. by guessing the right scenario, the guesser only keeps her capital (and does not double it). Because of these less favourable default conditions, the equilibrium growth factor `'G'` is equal to the growth factor difference $\Delta G^*$ in the double or nothing setting.

````python
# Plot the heatmap
plt.figure(figsize=(8, 6))
ax = sns.heatmap(bayes_results_dict['G'],
                 cmap=cmap_dict['G'],
                 #cbar_kws={'label': r'$P$'},
                 square=True,
                 annot=False,
                 vmin=-np.log(4), vmax=0
                 )

# Plot the mask
#sns.heatmap(mask, mask=~mask, cmap=cmap_dict['cherry'], cbar=False, square=True, ax=ax)

# Configure the plot
plt.title(r'$\Delta G^*(K_A,K_B)$', fontsize=16, fontstyle='italic')
set_plot_attributes_finite(tick_positions=np.round(np.linspace(0, M, 6)).astype(int))
plt.show()
````

Reproducing plots from [Figure 17](https://arxiv.org/pdf/2402.15892#figure.caption.72) and [Figure 18](https://arxiv.org/pdf/2402.15892#figure.caption.73) in the paper, i.e. $p'^*_k(K_A,K_B)$ for $\mathfrak{BGame}(N = 4, K_A, K_B , M = 10)$ games, $k \in \{0, \dots, N\}$:

The absent splitting ratios $p'^*_k$ depend not only on $K_A$ and $K_B$ but on $k$ as well. This is why we introduce a $k$-dependent dictionary of masks:

````python
# Initialize mask_dict and bayes_p_prime_result_dict
mask_dict = {k: np.zeros((len(KA_range), len(KB_range)), dtype=bool) for k in range(N + 1)}
bayes_p_prime_result_dict = {k: np.empty((len(KA_range), len(KB_range)), dtype=object) for k in range(N + 1)}

# Populate the dictionaries
for i in KA_range:
    for j in KB_range:
        p_prime = bayes_results_matrix[i, j].get('p_prime', {})
        for k in range(N + 1):
            if k in p_prime:
                mask_dict[k][i, j] = False
                bayes_p_prime_result_dict[k][i, j] = p_prime[k]
            else:
                mask_dict[k][i, j] = True
                bayes_p_prime_result_dict[k][i, j] = np.nan
````

$p'^*_1(K_A,K_B)$:

````python
# Plot the heatmap
plt.figure(figsize=(8, 6))
ax = sns.heatmap(bayes_p_prime_result_dict[1].astype(float),
                 #cmap=cmap_dict['P'],
                 cmap='viridis',
                 square=True,
                 annot=False,
                 vmin=0, vmax=1
                 )

# Plot the mask
sns.heatmap(np.zeros_like(bayes_p_prime_result_dict[1], dtype=float),
            mask=~mask_dict[1],
            cmap=cmap_dict['cherry'],
            cbar=False,
            square=True,
            ax=ax)

# Configure the plot
plt.title(r"$p'^*_1(K_A,K_B)$", fontsize=16, fontstyle='italic')
set_plot_attributes_finite(tick_positions=np.round(np.linspace(0, M, 6)).astype(int))
plt.show()
````

## Binomial plots

### Binomial Fisher games

Binomial games are limiting cases of games with finite binary sequence sizes ($M$), where the length of the sequences goes to infinity ($M \to \infty$), while the density of 1-s in scenario A and B convergence to finite real numbers:

$$
K_A / M \to x_A \in \mathbb{R}, \quad K_B / M \to x_B \in \mathbb{R},
$$

Setting formatting and attributes for Binomial games:

````python
def set_plot_attributes_binom(ax, grid_size, num):

    # Set tick positions and labels
    tick_positions = np.linspace(0, grid_size - 1, num=6, dtype=int)
    tick_labels = np.round(np.linspace(0, 1, num=6), 2)

    ax.set_xticks(tick_positions)
    ax.set_xticklabels(tick_labels, rotation=0, fontsize=12)  # Set rotation to 0 degrees for x-ticks
    ax.set_yticks(tick_positions)
    ax.set_yticklabels(tick_labels, rotation=0, fontsize=12)

    ax.set_xlabel(r'$x_A$', fontsize=14, fontstyle='italic')
    ax.set_ylabel(r'$x_B$', fontsize=14, fontstyle='italic')

    ax.invert_yaxis()  # Reverse the direction of the y-axis
````

Computing all equilibrium values on a grid of $(x_A,x_B) \in [0,1]\times[0,1]$ values for Binomial Fisher games $\overline{\mathfrak{Game}}(N=4, x_A, x_B)$, and storing the results (grouped by equilibrium paramaters $P, k, \nu, s, v$) in `binom_results_dict`:

````python
# Set the sampling parameter
N = 2

# Set gridsize
grid_size = 500

# Generate the grid of x_A and x_B values
x_A_values = np.linspace(0, 1, grid_size)
x_B_values = np.linspace(0, 1, grid_size)

# Initialize a matrix to hold dictionaries with 'P', 'k', 'nu', 's', and 'v' values
binom_results_matrix = np.empty((len(x_A_values), len(x_B_values)), dtype=dict)

# Compute 'P', 'k', 'nu', 's', and 'v' values for each combination of KA and KB
for i, xA in enumerate(x_A_values):
    for j, xB in enumerate(x_B_values):
        result = fishergame_solve(N, [xA, xB])
        binom_results_matrix[i, j] = result

# Keys for results_dict
keys = list(binom_results_matrix[0, 0].keys())

# Create a dictionary of matrices
binom_results_dict = {key: np.zeros((len(x_A_values), len(x_B_values)), dtype=float) for key in keys}

# Populate the dictionary of matrices
for i, xA in enumerate(x_A_values):
    for j, xB in enumerate(x_B_values):
        for key in keys:
            binom_results_dict[key][i, j] = binom_results_matrix[i, j][key]
````

## Prior Approximation

Binomial Bayesian limiting prior approximation based on equation ([174](https://arxiv.org/pdf/2402.15892#equation.4.174)):

$$
P^\approx_\text{⌘} = \frac{ \log \left ( \frac{(1-x_0^*) x_B}{(1-x_B)x_0^*}  \right )}{\log \left ( \frac{(1-x_A) x_B}{(1-x_B) x_A} \right )},
\quad
x_0^*(x_A,x_B) = \frac{\log \left ( \frac{1-x_A}{1-x_B} \right )}{\log \left ( \frac{(1-x_A) x_B}{(1-x_B) x_A} \right )}
$$

````python
# Zero'th order prior approximation formula
def prior_approx_zero(xA, xB):

    if xA == 0 or xA == 1 or xB == 0 or xB == 1:
      return {"P": None}

    if xA == xB:
      return {"P": 0.5}

    x0 = np.log((1 - xA) / (1 - xB)) / np.log((1 - xA) / (1 - xB) * xB / xA)
    P = np.log((1 - x0) / (1 - xB) * xB / x0) / np.log((1 - xA) / (1 - xB) * xB / xA)
    #cth = np.log(np.log(((1 - x0) * xB) / (x0 * (1 - xB))) / np.log((x0 * (1 - xA)) / ((1 - x0) * xA)))

    res = {"P": P}

    return res
````

$P^\approx_\text{⌘}(x_A,x_B)$:

````python
#from mpl_toolkits.axes_grid1 import make_axes_locatable

# Generate the grid of x_A and x_B values
grid_size = 400
x_A_values = np.linspace(0, 1, grid_size)
x_B_values = np.linspace(0, 1, grid_size)
X_A, X_B = np.meshgrid(x_A_values, x_B_values)

# Initialize the result matrix
results_P = np.zeros((grid_size, grid_size))

# Solve the game for each pair of x_A and x_B
for i in range(grid_size):
    for j in range(grid_size):
        result = prior_approx_zero(X_A[i, j], X_B[i, j])
        results_P[i, j] = result['P']

# Plot the contour plot
fig, ax = plt.subplots(figsize=(8, 8))  # Adjust the figure size to be square
contour_levels = np.linspace(0, 1, 101)

# Creating the filled contour plot
contourf = ax.contourf(X_A, X_B, results_P, levels=contour_levels, cmap=cmap_dict['P'])

# Adding contour lines
contour = ax.contour(X_A, X_B, results_P, levels=contour_levels, colors='black', linewidths=0.5)

# Create an axis for the color bar
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.35)

# Add a color bar
cbar = plt.colorbar(contourf, cax=cax)
cbar.outline.set_edgecolor(None)  # Remove the black boxing around the color bar

# Configure the plot
ax.set_title(r'$P^\approx_\text{⌘}(x_A,x_B)$', fontsize=16, fontstyle='italic')
set_plot_attributes_binom_contour(ax, grid_size=grid_size, num=6)


plt.show()
````

## Statistical Games and Unification

Specific colours and colour blending function for the reproduction

````python
def blend_colors(color1, color2, blend_factor):
    """Blend two colors by the specified blend factor."""
    return tuple(np.array(color1) * (1 - blend_factor) + np.array(color2) * blend_factor)

# Define colors
green = (0, 170 / 255, 0)
blue = (0, 0, 170 / 255)
````

Reproducing [Figure 39](https://arxiv.org/pdf/2402.15892#figure.caption.114):

$p'^*_{\gamma,k}$ for $\mathfrak{SGame}(N=4,K_A=5,K_B=8,M=14, \gamma)$ game as a function of $\gamma \in [0.01, 3]$:

````python
# Set the parameters
N = 4
M = 14

KA = 5
KB = 8

# Define the range for gamma
gamma_values = np.linspace(0.01, 3, 80)

# Initialize data storage
k_values = range(N+1)  # Assuming N = 4 for simplicity
p_prime_data = {k: [] for k in k_values}

# Calculate p_prime for each gamma
for gamma in gamma_values:
    result = statisticalgame_solve(N, [KA, KB], M, gamma=gamma, max_error=0.00001)
    for k in k_values:
        p_prime_data[k].append(result['p_prime'][k])

# Set up the plot
plt.figure(figsize=(9, 6))
plt.grid(False)

# Plot each k line with blended colors
for k in k_values:
    blend_factor = k / N
    color = blend_colors(green, blue, blend_factor)
    label = r"$p'^*_{\gamma,k=" + str(k) + "}$"
    sns.lineplot(x=gamma_values, y=p_prime_data[k], color=color, linewidth=2.5, label=r"$p'^*_{\gamma,k=" + str(k) + "}$")

# Add gridlines
plt.axvline(x=1, color='grey', linestyle='-', linewidth=0.5)
plt.axhline(y=0.5, color='grey', linestyle='-', linewidth=0.5)
plt.axhline(y=14/39, color='grey', linestyle='-', linewidth=0.5)

# Set tick labels with specific font size and no rotation
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12, rotation=0)

# Set plot labels and title
plt.xlabel(r'$\gamma$', fontsize=14, fontstyle='italic')
plt.ylabel(r"$p'_k$", fontsize=14, fontstyle='italic')
plt.title(r"$p'^*_{\gamma, k}$ as a function of $\gamma$", fontsize=16, fontstyle='italic')
#plt.legend(title='Values of k')
plt.xlim(0, 3)  # Assuming \gamma values are between 0 and 3
plt.ylim(0, 1.01)  # Assuming p_prime values are between 0 and 1

# Remove top and right spines using plt
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Move legend to the right side
plt.legend(bbox_to_anchor=(1.05, 0.71), loc='upper left', fontsize=12)

# Show the plot
plt.show()
````
